```python
import pandas as pd
#fetching raw data file
frailty_raw_data=pd.read_csv('/content/frailty_dataset.csv')
```

```python
#printing raw data
frailty_raw_data
```

```python
#Preprocessing the data to perform the analysis
#updating the column names with the related units of measure
frailty_raw_data.columns=['Height(in)', 'Weight(lb)', 'Age', 'Grip_strength(kg)','Frailty']
frailty_raw_data.columns

# Changing the categorical values into numerical class values
frailty_raw_data['Frailty'].replace(['Y', 'N'],
                        [0, 1], inplace=True)
```

```python
frailty_clean_data=frailty_raw_data.dropna()
frailty_clean_data
```

```python

frailty_clean_data.to_csv('/content/data_clean/frailty_clean_data.csv', index=False)

print("File saved successfully.")
```
