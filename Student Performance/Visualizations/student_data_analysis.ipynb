```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df=pd.read_csv('/content/data_clean')
df
```

```python
#Visualisation1

# Lets find the gender distribution among the dataset we gathered.

plt.bar(df['gender'].value_counts().index,
       df['gender'].value_counts().values,
       color = ['green', 'blue'])
plt.title('Gender Distribution')
plt.xticks(df['gender'].value_counts().index,['Female','Male'])
plt.savefig('/content/results')
```

```python
#Visualisation 2

# different scores distribution based on gender
figure, axis = plt.subplots(2, 2)


sns.boxplot(x=df['gender'],y=df['math score'],ax=axis[0,0])
sns.boxplot(x=df['gender'],y=df['reading score'],ax=axis[0,1])
sns.boxplot(x=df['gender'],y=df['writing score'],ax=axis[1,0])
sns.boxplot(x=df['gender'],y=df['Total score'],ax=axis[1,1])
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Scores distribution by gender')
plt.savefig('/content/results')
```

```python
#Visualisation 3

#Graph representing correlation between math score and reading score of male & female
plt.scatter(df['math score'], df['reading score'], c=df['gender'])
plt.legend(['female'])
plt.xlabel('math score')
plt.ylabel('reading score')
plt.title('math score distribution vs reading score distribution')
plt.savefig('Scores distribution by gender')
```

```python
#visualisation 4

figure, axis = plt.subplots(2,2)

sns.histplot(df['math score'],ax=axis[0,0],label='math score')
sns.histplot(df['reading score'],ax=axis[0,1],label='reading score')
sns.histplot(df['writing score'],ax=axis[1,0],label='writing score')
sns.histplot(df['Total score'],ax=axis[1,1],label='Total score')
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Frequency distribution of scores')
plt.savefig('/content/results')
```

```python
#Visualisation5

graph = sns.FacetGrid(df, col ='gender',hue_kws={'color': ['r', 'b']}, hue ='test preparation course')
# map the above form facetgrid with some attributes
graph.map(plt.hist, 'race/ethnicity', bins = 15, color ='orange').add_legend()
# show the object
plt.savefig('/content/results')
```
