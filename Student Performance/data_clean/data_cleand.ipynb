```python
import pandas as pd
```

```python
#fetching raw data file
student_raw_data=pd.read_csv('/content/StudentsPerformance.csv')
student_raw_data
```

```python
#Preprocessing the data to perform the analysis
# Changing the categorical values into numerical class values
student_raw_data['gender'].replace(['female', 'male'],
                        [0, 1], inplace=True)
student_raw_data['test preparation course'].replace(['none', 'completed'],
                        [0, 1], inplace=True)

#Creating a new column total score
student_raw_data['Total score'] = student_raw_data['math score'] + student_raw_data['writing score'] + student_raw_data['reading score']
student_raw_data
```

```python
#dropping of the rows if na occurances
student_clean_data=student_raw_data.dropna()
student_clean_data
#Saving the clean data into the csv file
# Save the cleaned data to a CSV file
student_clean_data.to_csv('/content/data_clean/cleaned_data.csv', index=False)
```
